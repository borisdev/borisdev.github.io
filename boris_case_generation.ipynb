```python
def get_new_events(synopsis: str, characters: list[dict]):
    data = dict(
        synopsis=synopsis,
        characters=characters
    )
    # TODO default to GPT4 with option parameter to GPT3 for speed
    event_endpoint = "https://causal-graph.casescope.com/events"

    r = requests.post(event_endpoint, json=data)
    print(r.status_code)
    data = r.json()
    return data

# Code to make timeline for evaluation of events

def events2df(events):
    character2events = {}

    for event in events:
        characters = event["characters"]
        for character in characters:
            if character not in character2events:
                character2events[character] = {
                    "character": character,
                    -6: "",
                    -5: "",
                    -4: "",
                    -3: "",
                    -2: "",
                    -1: "",
                    0: "",
                }
            era = event["era"]
            character2events[character][era] = (
                event["name"] + "\n\n" + event["description"] + "\n\n" + event["job"]
            )

    # print(character2events)

    df = pd.DataFrame(character2events.values())
    df = df[["character", -6, -5, -4, -3, -2, -1, 0]]
    return df

def pretty_print(df):
    return display(HTML(df.to_html().replace("\\n", "<br>")))

def events2timeline(events):
    df = events2df(events)
    pretty_print(df)

def oldevents2df(events):
    character2events = {}
    eras = []

    for event in events:
        character = event["charactersOnCasesUid"]

        if character not in character2events:
            character2events[character] = {"character": character}
        era = event["era"]
        eras.append(era)
        character2events[character][era] = event["name"] + "\n\n" + event["description"]

    # print(character2events)

    df = pd.DataFrame(character2events.values())
    sorted_eras = sorted(set(eras))
    df = df[["character"] + sorted_eras]
    return df

def oldevents2timeline(events):
    df = oldevents2df(events)
    pretty_print(df)
```

```python
# Evaluation

def events2df(events):
    character2events = {}

    for event in events:
        characters = event["characters"]
        for character in characters:
            if character not in character2events:
                character2events[character] = {
                    "character": character,
                    -6: "",
                    -5: "",
                    -4: "",
                    -3: "",
                    -2: "",
                    -1: "",
                    0: "",
                }
            era = event["era"]
            character2events[character][era] = (
                event["name"] + "\n\n" + event["description"] + "\n\n" + event["job"]
            )

    # print(character2events)

    df = pd.DataFrame(character2events.values())
    df = df[["character", -6, -5, -4, -3, -2, -1, 0]]
    return df

def pretty_print(df):
    return display(HTML(df.to_html().replace("\\n", "<br>")))

def events2timeline(events):
    df = events2df(events)
    pretty_print(df)

events2timeline(events)
```
